Случайная величина задана плотностью распределения, найти коэффициент а:

$$ f(x) = \begin{cases}
a \cdot \cos(x) & \text{при } -\frac{\pi}{2} < x \leq \frac{\pi}{2} \\
0 & \text{в остальных случаях}

```python
x = np.linspace(-np.pi / 2, np.pi / 2, 50)
x1 = np.linspace(-3, -np.pi / 2, 50)
x2 = np.linspace(np.pi / 2, 3, 50)

y_cos = np.cos(x)
y_zero = np.zeros(50)

plt.plot(x, y_cos)
plt.plot(x1, y_zero)
plt.plot(x2, y_zero)
```

$$\textbf{Задача 4}$$

 Поезда в метро следуют с интервалом 2 минуты. Пассажир в случайный момент времени приходит на платформу и ожидает ближайший поезд. Пусть X – время ожидания пассажира. Найти для X функцию распределения, математическое ожидание и дисперсию.

Функция распределения случайно величины x будет выглядеть так:
$$
F(x) =
\begin{cases}
0, & \text{если } x < 0 \ \\
\frac{x}{2}, & \text{если } 0 \leq x < 2 \ \\
1, & \text{если } x \geq 2
\end{cases} \\
\text{Математическое ожидание можно найти по формуле: } \\
E[X] = \int_{-\infty}^{\infty} x \cdot f(x) , dx \\
E[X] = \int_{0}^{2} x \cdot f(x) , dx = \int_{0}^{2} x \cdot \frac{1}{2} , dx = \frac{1}{2} \cdot \frac{x^2}{2} \Big|_{0}^{2} = \frac{1}{2} \cdot 2 = 1 \\
\text {Итого, }\textbf{математическое ожидание }\text{времени ожидания будет равно одной минуте.}\\
\text{Дисперсия вычисляется по формуле: } \\
Var(X) = E[X^2] - (E[X])^2 \\
E[X^2] = \int_{0}^{2} x^2 \cdot \frac{1}{2} , dx = \frac{1}{2} \cdot \frac{x^3}{3} \Big|_{0}^{2} = \frac{1}{2} \cdot \frac{8}{3} = \frac{4}{3} \\
Var(X) = E[X^2] - (E[X])^2 = \frac{4}{3} - 1 = \frac{1}{3} \\
\text{Итого, } \textbf{дисперсия } \text{будет равна } \frac{1}{3}
$$

\\

\text{Функция Гаусса (функция плотности нормального распределения:} \\
 f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}

$$

```python
from scipy.stats import norm
mu = 7
sigma = 16 ** 0.5

x15 = norm.cdf(x = 15, loc = mu, scale = sigma)
print('Вероятность P(x <= 15) = ', x15)
```

$$
Z = \frac{15 - 7}{\sqrt{16}} = \frac{8}{4} = 2  \\
P(Z \leq 2 ) \sim 0.977
$$
